```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# read epldata
data = pd.read_csv('epldata.csv')
# print first 5 rows
data.head()
```

```python
# inspect the data
data.info()
print(data.columns)
```

```python
# summary statistics
print(data.describe())
print('The oldest player is ', data.loc[:, ['Age']].max(), 'years old.')
print('The youngest player is ', data.loc[:, ['Age']].min(), 'years old.')
```

```python
# use ggplot for prettier plots
plt.style.use('ggplot')
# put plots to pdf
plt.savefig('output.pdf')
```

```python
# create a figure and axis
fig, ax = plt.subplots()

# Scatterplot of Pace vs Agility
ax.scatter(data['Pace'], data['Agility'])
# set a title and labels
ax.set_title('Pace vs Agility')
ax.set_xlabel('Pace')
ax.set_ylabel('Agility')
```

```python
# create a figure and axis 
fig, ax = plt.subplots() 
# count the occurrence of each class 
data1 = data['Foot'].value_counts() 

# get x and y data 
points = data1.index 
frequency = data1.values 

# create bar chart 
ax.bar(points, frequency) 

# set title and labels 
ax.set_title('Right vs Left Strong Feet') 
ax.set_xlabel('Foot') 
ax.set_ylabel('Number of Players')

print('The distribution of Players strong feet is;\n', data1.head())
```

```python
# Select a few columns from the dataset
subset = data[['Age', 'BodyType', 'FIFA Wage', 'Market Value in Euros', 'Dribbling', 'Shooting', 'Defending', 'Physical']]

# get columns to plot
data2 = subset.columns.drop(['Age', 'BodyType', 'FIFA Wage', 'Market Value in Euros'])
# create x data
x_data = range(0, subset.shape[0])
# create figure and axis
fig, ax = plt.subplots()
# plot each column
for column in data2:
    ax.plot(x_data, subset[column], label=column)
# set title and legend
ax.set_title('EPL Dataset Trends')
ax.legend()
```

```python
subset = data[['Age', 'BodyType', 'FIFA Wage', 'Market Value in Euros', 'Dribbling', 'Shooting', 'Defending', 'Physical']]

# get columns to plot
columns = subset.drop(['Age', 'BodyType', 'FIFA Wage', 'Market Value in Euros', 'Dribbling', 'Shooting', 'Defending'], axis=1)
plots = columns.plot.line(title='EPL Dataset Trends')
plots
```
